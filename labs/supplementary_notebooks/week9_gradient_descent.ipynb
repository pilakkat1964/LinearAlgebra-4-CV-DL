# Gradient Descent (Week 9)

## Objective
Visualize optimization paths over 3D surfaces to understand convergence in optimization routines.

```sagemath
%%sage
# Example: Himmelblau's Function
x, y = var('x y')
z = (x^2 + y - 11)^2 + (x + y^2 - 7)^2
contour_plot(z, (x, -5, 5), (y, -5, 5))
```
