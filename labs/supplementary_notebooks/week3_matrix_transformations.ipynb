# Matrix Transformations (Week 3)

## Objective
Explore and understand the effects of various matrix operations, such as reflection, rotation, and scaling, on vectors and graphs using SageMath.

```sagemath
%%sage
# Example: Reflection Matrix
M = matrix([[1, 0], [0, -1]])
v = vector([1, 1])
plot(M*v, color='blue') + plot(v, color='red')
```
